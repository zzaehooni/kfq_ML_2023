```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
data = cancer.data
label = cancer.target
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df.columns = cancer.feature_names
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(data)
scaled_data = ss.transform(data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(scaled_data,label)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
lr = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(n_estimators=100)
ada = AdaBoostClassifier(n_estimators=100)
```

```python
models = [lr,knn,dt,rf,ada]
meta_train =[]
meta_test=[]
df_train = pd.DataFrame()
df_test = pd.DataFrame()
for i, model in enumerate(models):
  model.fit(x_train,y_train)
  df_train[str(i)] = model.predict(x_train)
  df_test[str(i)] = model.predict(x_test)
```

```python
lr = LogisticRegression()
lr.fit(df_train,y_train)
pred = lr.predict(df_test)
pp = lr.predict_proba(df_test)[:,1]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score

def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix( y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test, pred_proba)
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
get_clf_eval(y_test,pred,pp)
```

```python
lr = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier(n_estimators=100)
ada = AdaBoostClassifier(n_estimators=100)
models = [lr,knn,dt,rf,ada]
for i, model in enumerate(models):
  model.fit(x_train,y_train)
  pred = model.predict(x_test)
  pp = model.predict_proba(x_test)[:,1]
  get_clf_eval(y_test,pred,pp)
```
